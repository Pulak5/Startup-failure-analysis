```python
base_url="https://www.failory.com"
```

```python
def get_data(base_url,ts,te,list,records):
    for c in list:
        url=base_url+ts+c+te
        time.sleep(1)
        res=requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
        soup=BeautifulSoup(res.content,"html.parser")
        blocks=soup.find_all("div",class_="listicle-startup-collection-item")
        print(len(blocks))
        for b in blocks:
            text=b.get_text(separator="|",strip=True)
            parts=text.split("|")
            data = {
                'Startup': parts[0],
                'Description': parts[1]
            }

            for i in range(len(parts)-1):
                key = parts[i].strip().lower()

                if 'founder' in key:
                    data['Founders'] = parts[i+1]

                elif key == 'country:':
                    data['Country'] = parts[i+1]

                elif key == 'industry:':
                    data['Industry'] = parts[i+1]

                elif key == 'started in:':
                    data['Started in'] = parts[i+1]

                elif key == 'closed in:':
                    data['Closed in'] = parts[i+1]

                elif 'employees' in key:
                    data['Employees'] = parts[i+1]

                elif key == 'funding amount:':
                    data['Funding Amount'] = parts[i+1]

                elif 'specific' in key:
                    data['Failure Cause'] = parts[i+1]

            records.append(data)
```

```python
countries=["germany","india","belgium","russia","bulgaria","spain","sweden","canada"]
categories=["fashion","education","artificial-intelligence","e-commerce","crypto","foodtech","depending-on-others","productivity","health","sports","transportation"]
records=[]
ts="/startups/"
te="-failures"
get_data(base_url,ts,te,countries,records)
print(len(records))
```

```python
get_data(base_url,ts,te,categories,records)
```

```python
df=pd.DataFrame(records)
df['Employees'] = df['Employees'].astype(str)
```
